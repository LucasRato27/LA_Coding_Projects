# BigQuery Starter Notebook
Connected to project: `insider_data_lake`

```python
from google.cloud import bigquery
import pandas as pd

PROJECT_ID = 'insider_data_lake'
client = bigquery.Client(project=PROJECT_ID)
print('Connected to BigQuery project:', PROJECT_ID)
```

## List available datasets

```python
datasets = list(client.list_datasets())
for ds in datasets:
    print(ds.dataset_id)
```

## Run a query

```python
query = """
    SELECT *
    FROM `insider_data_lake.DATASET.TABLE`
    LIMIT 10
"""

df = client.query(query).to_dataframe()
df.head()
```
